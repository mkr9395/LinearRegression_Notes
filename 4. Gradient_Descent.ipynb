## Functions

A function is a mathematical rule that takes an input value, processes it according to a specific
formula or set of instructions, and produces a unique output value. In other words, a function
is a relationship between input and output values where each input is connected to exactly
one output.

### Parameters in a Function:

- In mathematics, parameters of a function are the variables that are used to define the
behaviour of the function. The parameters influence the function's output by determining how
the input values are processed.


- The parameters are the constants or coefficients that appear in the function's formula. 



- For example, in the quadratic function $f(x) = ax^2 + bx + c$, 



'a', 'b', and 'c' are the parameters of the
function. By changing the values of these parameters, you can modify the shape and position
of the parabola represented by the function.

### Parametric vs Non-parametric ML models:

Parametric and non-parametric are two different approaches to machine learning models, and they differ in how they make assumptions about the underlying data distribution and the number of parameters they use.

1. **Parametric Models:**
   
   - Parametric models make strong assumptions about the functional form of the underlying data distribution.<br></br>
   
   - They have a fixed number of parameters that are determined by the model structure and do not depend on the amount of training data.<br></br>
   
   - Examples of parametric models include linear regression, logistic regression, and Naive Bayes. In linear regression, for instance, the model assumes a linear relationship between the input features and the output, and the number of parameters depends on the number of features.



2. **Non-Parametric Models:**
   
   - Non-parametric models do not make strong assumptions about the underlying data distribution and can be more flexible in capturing complex patterns.<br></br>
   - They can adapt to the complexity of the data by using a variable number of parameters, which may grow with the amount of training data.<br></br>  
   - Examples of non-parametric models include k-nearest neighbors (KNN), decision trees, random forests, support vector machines (SVM), and various types of neural networks. KNN, for instance, does not assume any specific functional form for the data distribution and can adapt to different data patterns.
<br></br>


In summary:


- Parametric models make strict assumptions about the data distribution and have a fixed number of parameters<br></br>

- Non-parametric models make fewer assumptions about the data distribution and can adapt to varying complexities with a flexible number of parameters.<br></br>

- The choice between parametric and non-parametric models depends on the nature of the data and the problem you are trying to solve. Parametric models may work well when the data conforms to the assumptions, while non-parametric models can be more suitable for complex or non-linear relationships.

### Loss function:

A loss function, also known as a cost function or objective function, is a mathematical function
that measures the difference between the predicted output and the actual target values in a
machine learning model. 


The primary goal of training a machine learning model is to minimize
the value of the loss function, which corresponds to improving the model's performance on
the given task.


Loss functions play a crucial role in the optimization process, guiding the learning algorithm to
adjust the model's parameters to achieve better predictions.

### How to select a good Loss Function:

1. __Problem type :__ The choice of a loss function depends on the type of problem you are
solving. For example, in regression tasks, mean squared error (MSE) or mean absolute error
(MAE) are commonly used. For binary classification, cross-entropy loss or hinge loss can be
employed. For multi-class classification, categorical cross-entropy or multi-class hinge loss
can be used. Choose a loss function that aligns with the objectives of the specific problem
you are addressing.



2. __Robustness to outliers :__ Some loss functions, like mean squared error, are more sensitive to
outliers, which can lead to a model that is overly influenced by extreme values. If your
dataset contains outliers or is prone to noise, consider using a loss function that is more
robust to outliers, such as mean absolute error (MAE) or Huber loss.


3. __Interpretability and ease of use :__ A good loss function should be interpretable and easy to
implement. Simple loss functions like mean squared error or cross-entropy loss are widely
used because they are easy to understand, compute, and differentiate. When possible, opt
for a loss function that is easy to work with and can be easily incorporated into your
optimization process.


4. __Differentiability :__ Most optimization algorithms, like gradient descent, require the loss
function to be differentiable. Choose a loss function that has continuous first-order
derivatives, which makes it easier to compute the gradients needed for optimization.


5. __Compatibility with the model :__ Ensure that the chosen loss function is compatible with the
model architecture you are using. Some models have specific requirements or assumptions
about the loss function. For example, linear regression assumes a Gaussian noise distribution, which is why mean squared error is a suitable loss function in that case.

### Problems with easy way $\big(\frac{dL}{g\beta_0}or\;slope=0\big)$

1. __Non-convexity :__ The loss function may not always be convex, meaning that it might have
multiple local minima and maxima. In such cases, setting the gradient to zero might lead to
a local minimum or maximum, which is not necessarily the global minimum (the optimal
solution).


2. __Complexity :__ For some models, the loss function can be highly complex, and finding the
analytical solution by setting the gradient to zero might be computationally expensive or
even impossible. This is particularly true for deep learning models, where the loss functions
involve a large number of parameters and complex relationships between them.


3. __Scalability :__ In large-scale machine learning problems with massive amounts of data or high-
dimensional feature spaces, computing the analytical solution by setting the gradient to
zero can be computationally prohibitive due to the high cost of processing and storing the
data.


4. __Online learning and streaming data :__ In some applications, the data is not available all at
once but arrives in a continuous stream. In these scenarios, models need to be updated
incrementally as new data arrives, and an analytical solution would not be practical.
Gradient descent and its variants, such as stochastic gradient descent, are well-suited for
online learning and handling streaming data.

### Convex and Non-convex functions:

- **Convex Functions:**<br></br>
    - These functions have a U-shaped graph, and any line segment connecting two points on the graph lies above the graph. They have a unique global minimum and are easy to optimize.<br></br>
    - A function $f(x)$ is considered convex over an interval if, for any two points $x_1$ and $x_2$ within that interval, the line segment between $x_1, f(x_1)$ and $x_2, f(x_2)$ lies above the graph of the function.<br></br>
    - Convex functions have only one minimum point, which is also the global minimum.

<br></br>

- **Non-Convex Functions:**<br></br>
    - These functions have irregular or complex shapes, with line segments that may not always lie above the graph. They can have multiple local minima, making optimization more challenging.<br></br>
    - Non-convex functions can have multiple local minima and maxima, making them more complex to optimize.<br></br>
    - The presence of non-convexity can pose challenges in optimization problems because traditional optimization algorithms may not guarantee finding the global minimum.

Convex functions are desirable in optimization because they have a unique global minimum, and optimization algorithms can reliably find this minimum. In contrast, non-convex functions can be challenging to optimize, as they may have multiple local optima, making it difficult to determine the best solution.

# Gardient Descent

- It is an Optimization algorithm which gives minima of a differentiable function (differentiable at all points)



- The idea is to take repeated steps in the opposite direction of the gradient (or approximate gradient) of the function at the current point, because this is the direction of steepest descent.

sklearn uses OLS

if slope = -ve  increment in c

if slope = +ve  decrement in c

$$\Longrightarrow m_{new}\;=\;m_{old}\;-\;\eta\;*\;\;\frac{1}{n}\; * \text{loss_slope}$$

$$\text{where, loss_slope}\;\Big(\frac{\delta L}{\delta m}\Big)\;\;=\;-2 * \sum_{i=1}^{n}\;(y - mx - c) * x$$

<br></br>
$$\Longrightarrow c_{new}\;=\;c_{old}\;-\;\eta\;*\;\;\frac{1}{n}\;* \text{loss_intercept}$$

$$\text{loss_intercept} \;\Big(\frac{\delta L}{\delta c}\Big)\;=\;-2 * \;\sum_{i=1}^{n}(y - mx - c)$$



$where\;\;\eta*loss\;\longrightarrow\;step-size$

The values will keep on updating as much the epochs is provided.

```python
class GDRegressor:
    
    def __init__(self,learning_rate,epochs):
        
        self.m = 27.82
        self.c = -120
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,X,y):
        # calculate c using Gradient descent
        
        for i in range(self.epochs):  # loop epoch times chalega
            loss_slope = -2 * np.sum(y - self.m*X.ravel() - self.c) # slope of loss function wrt slope
            
            # updating value of c each time inside loop
            # c_new = c_old - learning_rate * loss_slope 
            self.c = self.c - (self.lr * loss_slope) 
            
            
        print(self.c)
         
```

```python
class GDNewRegressor:
    
    def __init__(self,learning_rate,epochs):
        
        self.m = 100
        self.c = -120
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,X,y):
        
        #calculate the c using GD
        for i in range(self.epochs):
            
            loss_slope_m = -2 * np.sum((y - self.m*X.ravel() - self.m)*X.ravel()) # loss slope wrt m
            loss_slope_c = -2 * np.sum(y - self.m*X.ravel() - self.c)   # loss slope wrt c
            
            self.m = self.m - (self.lr * loss_slope_m)
            self.c = self.c - (self.lr * loss_slope_c)
            
    def predict(self,X):
        return self.m * X + self.c
```

```python
b = -520
m = 600
lr = 0.001
all_b = []
all_m = []
all_cost = []

epochs = 30

for i in range(epochs):
    slope_b = 0
    slope_m = 0
    cost = 0
    for j in range(X.shape[0]):
        slope_b = slope_b - 2*(y[j] - (m * X[j]) -  b)
        slope_m = slope_m - 2*(y[j] - (m * X[j]) -  b)*X[j]
        cost = cost + (y[j] - m * X[j] -b) ** 2

    b = b - (lr * slope_b)
    m = m - (lr * slope_m)
    all_b.append(b)
    all_m.append(m)
    all_cost.append(cost)
```

## BATCH GRADIENT DESCENT

Batch Gradient Descent runs data for all value and then updates the m and c once.

It is not useful for very large dataset.

Helpful in Convex functions like Linear Regression but dataset should be small
.

-  __for a single epoch it calculate the Loss for Entire dataset and then updates the weights and baises.__



- In batch gradient descent, the entire training dataset is used to compute the gradient of the cost function. The model parameters are updated once for the entire dataset. It provides an accurate estimate of the gradient but can be computationally expensive for large datasets.


- number of epochs = number of updates

##### 10 times weight will be updated:



##### so for 10 epochs and 50 rows, total weights and biases update will be 10 updates only

#### for n-dimensional data


$$\text{Loss function (wrt any coefficients) for n-dim data}$$


$$\;\Longrightarrow\;\frac{dL}{d\beta_{m}}\;=\;\frac{-2}{n}*\sum_{i=1}^n\;(y_{i}\;-\;\hat y_{i})*x_{im}$$

$for\;\beta_0 = \frac{dL}{d\beta_{0}},\;for\;\beta_1 = \frac{dL}{d\beta_{1}},\;for\;\beta_2 = \frac{dL}{d\beta_{2}}\;$

where 

- $x_{i}$ is the total number of rows


- $L$ is the loss (typically the Mean Squared Error) that measures the error of your linear regression model.


-  $\frac{dL}{d\beta_m}$ represents the derivative of the loss with respect to the $m^{th}$ coefficient, $\beta_m$.


- n is total number of independent columns


- m is the total number of coefficents


- $x_{im}$ is particular coefficent's corresponding columns values. It is the value of the $i^{th}$ data point in the $m^{th}$ feature or coefficient's corresponding column.

It is a derivative of the Mean Squared Error (MSE) loss with respect to a particular coefficient, $\beta_m$, in a linear regression model.

### Creating Batch Gradient Descent from Scratch

```python
X_train.shape

# 10 columns (9 independent, 1 Dependent )
```

#### derivative of intercept : $$\frac{\delta L}{\delta \beta_0}=$$

`intercept_der = -2 * np.mean(y_train - y_hat)`



#### derivative of coefficient : 

`coef_der = -2 * np.dot((y_train - y_hat),X_train)/X_train.shape[0]`



#### value of $\hat y$

`y_hat = np.dot(X_train,self.coef_) + self.intercept`



```python
class BGDRegressor:
    
    def __init__(self,learning_rate=0.01,epochs=100):
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,X_train,y_train):
        # initialize your coefs
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        
        for i in range(self.epochs):
            # update all the coef and the intercept
            y_hat = np.dot(X_train,self.coef_) + self.intercept_ # known as VECTORIZATION
            #print("Shape of y_hat",y_hat.shape)
            
            intercept_der = -2 * np.mean(y_train - y_hat)
            self.intercept_ = self.intercept_ - (self.lr * intercept_der) # where intercept_der = derivate of loss ftn wrt to intercept(c)
            
            coef_der = -2 * np.dot((y_train - y_hat),X_train)/X_train.shape[0]
            self.coef_ = self.coef_ - (self.lr * coef_der)
        
        # print(self.intercept_,self.coef_)
    
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
class SGDRegressor:
    
    def __init__(self,learning_rate=0.01,epochs=100):
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,X_train,y_train):
        # init your coefs
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        
        for i in range(self.epochs):
            for j in range(X_train.shape[0]): # this loop will go on as many rows we have
                idx = np.random.randint(0,X_train.shape[0]) # here idx limit is from 0 to total number of rows of train
                
                y_hat = np.dot(X_train[idx],self.coef_) + self.intercept_
                
                intercept_der = -2 * (y_train[idx] - y_hat) # multiplying by idx row train i.e. train[idx]
                self.intercept_ = self.intercept_ - (self.lr * intercept_der)
                
                coef_der = -2 * np.dot((y_train[idx] - y_hat),X_train[idx])
                self.coef_ = self.coef_ - (self.lr * coef_der)
        
        print(self.intercept_,self.coef_)
    
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
sgd = SGDRegressor(learning_rate=0.01,epochs=40)
```

```python
start = time.time()
sgd.fit(X_train,y_train)
print("\nThe time taken is",time.time() - start)
```

```python
y_pred = sgd.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

## Mini Batch Gradient Descent

Uses batches for each epochs and updates per epochs.

Mini-batch gradient descent is a compromise between batch gradient descent and stochastic gradient descent. 


It computes the gradient and updates the model parameters using a small batch of training examples at a time. The batch size is typically chosen as a moderate number, providing a balance between accuracy and computational efficiency.

30 = rows in 1 batch


10 = number of batches


30 * 10 = 300 = total rows in dataset

### Steps:


- batches = 10 which is less than total number of rows


- for every epoch and entire 1 batch, Loss will be calculated altogether and weights will be updated



- this will be repeated as many batches are there for a single epoch, avg Loss will be calculated



- this will be repeated as many epochs will be there i.e number of update = number of epochs

### Mini Batch Gradient Descent from Scratch 

```python
import random

class MBGDRegressor:
    
    def __init__(self,batch_size,learning_rate=0.01,epochs=100): #asking batch size from user
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size
        
    def fit(self,X_train,y_train):
        # init your coefs
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        
        for i in range(self.epochs):
            
            for j in range(int(X_train.shape[0]/self.batch_size)):  # ander ka loop will run as much as number of batch is
                
                idx = random.sample(range(X_train.shape[0]),self.batch_size)
                
                y_hat = np.dot(X_train[idx],self.coef_) + self.intercept_
                #print("Shape of y_hat",y_hat.shape)
                intercept_der = -2 * np.mean(y_train[idx] - y_hat)
                self.intercept_ = self.intercept_ - (self.lr * intercept_der)

                coef_der = -2 * np.dot((y_train[idx] - y_hat),X_train[idx])
                self.coef_ = self.coef_ - (self.lr * coef_der)
        
        print(self.intercept_,self.coef_)
    
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
mbr = MBGDRegressor(batch_size=int(X_train.shape[0]/50),learning_rate=0.01,epochs=100)
```

```python
mbr.fit(X_train,y_train)
```

```python
y_pred = mbr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
